```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
var=sns.load_dataset('tips')
var
```

```python
sns.stripplot(x='day', y='total_bill', data=var)
plt.show()
```

```python
sns.boxplot(x='day', y='total_bill', data=var)
plt.show()
```

```python
sns.stripplot(x='day', y='total_bill', data=var, hue='sex')
plt.show()
```

```python
sns.stripplot(x='day', y='total_bill', data=var, hue='sex', palette='Set1_r')
plt.show()
```

```python
sns.stripplot(x='day', y='total_bill', data=var, hue='sex', palette='Set1_r', linewidth=0.5)
plt.show()
```

```python
sns.stripplot(x='day', y='total_bill', data=var, hue='sex', palette='Set1_r', linewidth=0.5, edgecolor='r')
plt.show()
```

```python
sns.stripplot(x='day', y='total_bill', data=var, hue='sex', palette='Set1_r',
              linewidth=0.5, edgecolor='r', jitter=2) # jitter for spreading the distribution
plt.show()
```

```python
sns.stripplot(x='day', y='total_bill', data=var, hue='sex', palette='Set1_r',
              linewidth=0.5, edgecolor='r', jitter=2, size=10) 
plt.show()
```

```python
sns.stripplot(x='day', y='total_bill', data=var, hue='sex', palette='Set1_r',
              linewidth=0.5, edgecolor='r', jitter=2, size=10,marker='<') 
plt.show()
```

```python
sns.stripplot(x='day', y='total_bill', data=var, hue='sex', palette='Set1_r',
              linewidth=0.5, edgecolor='r', jitter=2, size=10,marker='<', alpha=0.2) 
plt.show()
```

```python
sns.stripplot(x=var['total_bill'])
plt.show()
```
