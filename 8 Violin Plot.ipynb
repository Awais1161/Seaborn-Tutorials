```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
var=sns.load_dataset('tips')
var
```

```python
sns.violinplot(x='day', y='total_bill', data=var)
plt.show()
```

```python
sns.violinplot(x='day', y='total_bill', data=var,hue='time')
plt.show()
```

```python
sns.violinplot(x='day', y='total_bill', data=var,linewidth=5)
plt.show()
```

```python
sns.violinplot(x='day', y='total_bill', data=var,linewidth=5, palette='afmhot_r')
plt.show()
```

```python
sns.violinplot(x='time', y='total_bill', data=var)
plt.show()
```

```python
sns.violinplot(x='time', y='total_bill', data=var, order=['Dinner', 'Lunch'])
plt.show()
```

```python
sns.violinplot(x='time', y='total_bill', data=var, order=['Dinner', 'Lunch'], saturation=0.5)
plt.show()
```

```python
sns.violinplot(x='time', y='total_bill', data=var, color='r')
plt.show()
```

```python
sns.violinplot(x='day', y='total_bill', data=var, hue='sex')
plt.show()
```

```python
sns.violinplot(x='day', y='total_bill', data=var, hue='sex', split=True)
plt.show()
```

```python
sns.violinplot(x='day', y='total_bill', data=var, hue='sex', split=True, scale='count') # scale ki smj ni ai
plt.show()
```

```python
sns.violinplot(x='day', y='total_bill', data=var, hue='sex', split=True, scale='width')
plt.show()
```

```python
sns.violinplot(y='day', x='total_bill', data=var, hue='sex', split=True, scale='width')
plt.show()
```

```python
sns.violinplot(x='time', y='total_bill', data=var, inner='quart')
plt.show()
```

```python
sns.violinplot(x='time', y='total_bill', data=var, inner='box') # default
plt.show()
```

```python
sns.violinplot(x=var['total_bill']) # only for numeric values not for catagorical
plt.show()
```

```python
sns.violinplot(y=var['total_bill']) # only for numeric values not for catagorical
plt.show()
```
