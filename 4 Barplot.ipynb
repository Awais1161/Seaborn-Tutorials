```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
var=sns.load_dataset('penguins')
var
```

```python
sns.barplot(x=var.island, y=var.bill_length_mm)
plt.show()
```

```python
sns.barplot(x=var['island'], y=var['bill_length_mm'])    # same
plt.show()
```

```python
sns.barplot(x='island', y='bill_length_mm',data=var) # Same
plt.show()
```

```python
sns.barplot(x='island', y='bill_length_mm',data=var, hue='sex') 
plt.show()
```

```python
o=['Dream', 'Torgersen', 'Biscoe']
sns.barplot(x='island', y='bill_length_mm',data=var, hue='sex', order=o) 
plt.show()
```

```python
o=['Dream', 'Torgersen', 'Biscoe']
sns.barplot(x='island', y='bill_length_mm',data=var, hue='sex', order=o, hue_order=['Female', 'Male']) 
plt.show()
```

```python
o=['Dream', 'Torgersen', 'Biscoe']
sns.barplot(x='island', y='bill_length_mm',data=var, hue='sex'
            , order=o, hue_order=['Female', 'Male'],ci=100) 
plt.show()
```

```python
o=['Dream', 'Torgersen', 'Biscoe']
sns.barplot(x='island', y='bill_length_mm',data=var, hue='sex'
            , order=o, hue_order=['Female', 'Male'],ci=100, n_boot=100) # n_boot ki smj ni ai
plt.show()
```

```python
o=['Dream', 'Torgersen', 'Biscoe']
sns.barplot(x='island', y='bill_length_mm',data=var, hue='sex'
            , order=o, hue_order=['Female', 'Male'],ci=100, n_boot=100, orient='v') # by default vertical
plt.show()
```
