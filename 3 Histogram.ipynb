```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
var=sns.load_dataset('penguins')
var
```

```python
sns.displot(var['bill_length_mm'])
plt.show()
```

```python
sns.distplot(var['bill_length_mm'])
plt.show()
```

```python
sns.displot(var['flipper_length_mm'])
plt.show()
```

```python
sns.displot(var['flipper_length_mm'],bins=[170,180,190,200,210,220,230,240])
plt.show()
```

```python
sns.distplot(var['flipper_length_mm'],bins=[170,180,190,200,210,220,230,240])
plt.show()
```

```python
sns.displot(var['flipper_length_mm'],bins=[170,180,190,200,210,220,230,240],kde=True)
plt.show()
```

```python
sns.displot(var['flipper_length_mm'],bins=[170,180,190,200,210,220,230,240],kde=True,rug=True) #Problem
plt.show()
```

```python
sns.displot(var['flipper_length_mm'],bins=[170,180,190,200,210,220,230,240],kde=True,rug=True, color='red') 
plt.show()
```

```python
sns.displot(var['flipper_length_mm'],bins=[170,180,190,200,210,220,230,240],kde=True,rug=True, color='red', log_scale=True) 
plt.show()
```

```python
sns.displot(var['flipper_length_mm'],kde=True,rug=True, color='red', log_scale=True)  # without bins
plt.show()
```
