```python
import seaborn as sns 
import matplotlib.pyplot as plt

v=[1,2,3,4,5,6,7]
v1=[2,3,4,1,5,6,7]

plt.plot(v,v1)

plt.show()
```

```python
v=[1,2,3,4,5,6,7]
v1=[2,3,4,1,5,6,7]

sns.lineplot(x=v, y=v1) 

plt.show()   # this graph is 100 percent same as in matplotlib
```

```python
var=[1,2,3,4,5,6,7]
var_1=[2,3,4,1,5,6,7]

sns.lineplot(x='var', y='var_1')  # not work because seaborn work on dataset
```

```python
import pandas as pd

var=[1,2,3,4,5,6,7]
var_1=[2,3,4,1,5,6,7]

x_1=pd.DataFrame({'var':var, 'var_1':var_1})

sns.lineplot(x='var', y='var_1', data=x_1) 

plt.show()
```

```python
y_1=sns.load_dataset('penguins').head(20)
y_1
```

```python
sns.lineplot(x='bill_length_mm', y='flipper_length_mm', data=y_1,hue= 'sex',
             style='sex',palette='rocket',markers=['o','>'],
             dashes=False,legend=False) # jb k matplotlib ma legend lanay k lia legend ki zroorat thi

plt.show()
```

```python
sns.lineplot(x='bill_length_mm', y='flipper_length_mm', data=y_1,hue= 'sex',
             style='sex',palette='rocket',markers=['o','>'],
             dashes=False,legend='f') # koi b enter kr k error box ma all legends can be viwed

plt.show()
```

```python
sns.lineplot(x='bill_length_mm', y='flipper_length_mm', data=y_1,hue= 'sex',
             style='sex',palette='rocket',markers=['o','>'],
             dashes=False,legend='auto') 

plt.show()
```

```python
sns.lineplot(x='bill_length_mm', y='flipper_length_mm', data=y_1,hue= 'sex',
             style='sex',palette='rocket',markers=['o','>'],
             dashes=False,legend='auto') 

plt.grid()
plt.title('python')
plt.show()
```
