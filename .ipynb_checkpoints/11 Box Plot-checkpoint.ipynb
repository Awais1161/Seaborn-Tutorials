```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
var=sns.load_dataset('tips')
var
```

```python
sns.set(style='whitegrid')
sns.boxplot(x='day', y='total_bill', data=var)
plt.show()
```

```python
sns.set(style='whitegrid')
sns.boxplot(x='day', y='total_bill', data=var, hue='time')
plt.show()
```

```python
sns.set(style='whitegrid')
sns.boxplot(x='day', y='total_bill', data=var, hue='sex')
plt.show()
```

```python
sns.set(style='whitegrid')
sns.boxplot(x='day', y='total_bill', data=var, hue='sex',color='g')
plt.show()
```

```python
sns.set(style='whitegrid')
sns.boxplot(x='day', y='total_bill', data=var, hue='sex',color='g', order=['Fri', 'Sun', 'Thur', 'Sat'])
plt.show()
```

```python
sns.set(style='whitegrid')
sns.boxplot(x='day', y='total_bill', data=var, hue='sex',
            color='r', order=['Fri', 'Sun', 'Thur', 'Sat'], showmeans=True) # arrows show means
plt.show()
```

```python
sns.set(style='whitegrid')
sns.boxplot(x='day', y='total_bill', data=var, hue='sex',
            color='r', order=['Fri', 'Sun', 'Thur', 'Sat'], showmeans=True, meanprops={'marker':'+', 'markeredgecolor':'b'}) 
plt.show()
```

```python
sns.set(style='whitegrid')
sns.boxplot(x='day', y='total_bill', data=var, hue='sex',
            color='r', order=['Fri', 'Sun', 'Thur', 'Sat'], 
            showmeans=True, meanprops={'marker':'+', 'markeredgecolor':'b'},linewidth=5) 
plt.show()
```

```python
sns.set(style='whitegrid')
sns.boxplot(x='day', y='total_bill', data=var, hue='sex',
             order=['Fri', 'Sun', 'Thur', 'Sat'], 
            showmeans=True, meanprops={'marker':'+', 'markeredgecolor':'b'},linewidth=5, palette='Spectral_r') 
plt.show()
```

```python
sns.set(style='whitegrid')
sns.boxplot(x='day', y='total_bill', data=var, hue='sex',
             order=['Fri', 'Sun', 'Thur', 'Sat'], 
            showmeans=True, meanprops={'marker':'+', 'markeredgecolor':'b'},linewidth=5, palette='Spectral_r', orient='v') 
plt.show()                                                                                             # vertical by drfault
                                                                  # for horizontal x,y,should both be numerical
                                                                   # ya phr nichay wali command
```

```python
sns.set(style='whitegrid')
sns.boxplot(x='total_bill', y='day', data=var, hue='sex',
             order=['Fri', 'Sun', 'Thur', 'Sat'], 
            showmeans=True, meanprops={'marker':'+', 'markeredgecolor':'b'},linewidth=5, palette='Spectral_r', orient='v') 
plt.show()                                                                                            
              # run huna chaye tha
```

```python
sns.set(style='whitegrid')
sns.boxplot(data=var,  
            showmeans=True, meanprops={'marker':'+', 'markeredgecolor':'b'},linewidth=5, palette='Spectral_r', orient='h') 
plt.show()              # for horizontal (especial case)
```
