```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
var=sns.load_dataset('tips')
var
```

```python
sns.boxplot(y=var['total_bill'])
plt.show()
```

```python
sns.barplot(x='day', y='total_bill', data=var)
plt.show()
```

```python
sns.barplot(x='day', y='total_bill', data=var)
plt.grid()
plt.show()
```

```python
sns.set_style('white')   # no change bcz of by default
sns.barplot(x='day', y='total_bill', data=var) 

plt.show()
```

```python
sns.set_style('dark')    # background
sns.barplot(x='day', y='total_bill', data=var) 

plt.show()
```

```python
sns.set_style('whitegrid')    
sns.barplot(x='day', y='total_bill', data=var) 

plt.show()
```

```python
sns.set_style('darkgrid')    
sns.barplot(x='day', y='total_bill', data=var) 

plt.show()
```

```python
sns.set_style('ticks')  # default  
sns.barplot(x='day', y='total_bill', data=var) 

plt.show()
```

```python
sns.set_style('darkgrid')    
sns.barplot(x='day', y='total_bill', data=var) 
plt.grid()
plt.show()
```

```python
sns.set_style('white')    
sns.barplot(x='day', y='total_bill', data=var) 

plt.show()
```

```python
sns.set_style('white')    
sns.barplot(x='day', y='total_bill', data=var) 
sns.despine()
plt.show()
```

```python
sns.set_style('white') 
plt.figure(figsize=(3,12))  # width:height
sns.barplot(x='day', y='total_bill', data=var) 
sns.despine()
plt.show()
```

```python
sns.set_style('white') 
plt.figure(figsize=(12,3))
sns.barplot(x='day', y='total_bill', data=var) 
sns.despine()
plt.show()
```

```python
sns.set_style('white') 
plt.figure(figsize=(12,15)) 
sns.barplot(x='day', y='total_bill', data=var) 
sns.despine()
plt.show()
```

```python
sns.set_style('white') 
sns.set_context('poster') # poster means a style(many kinds k styles hn)
sns.barplot(x='day', y='total_bill', data=var) 

plt.show()
```

```python
sns.set_style('white') 
sns.set_context('paper') 
sns.barplot(x='day', y='total_bill', data=var) 

plt.show()
```

```python
sns.set_style('white') 
sns.set_context('poster',font_scale=2)
sns.barplot(x='day', y='total_bill', data=var) 

plt.show()
```

```python
sns.set_style('white') 
sns.set_context('paper') 
sns.barplot(x='day', y='total_bill', data=var ,color='r') 

plt.show()
```
