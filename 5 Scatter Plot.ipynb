```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
var=sns.load_dataset('penguins')
var
```

```python
sns.scatterplot(x='bill_length_mm', y='bill_depth_mm',data=var)
plt.show()
```

```python
sns.scatterplot(x='bill_length_mm', y='bill_depth_mm',data=var, hue='sex')
plt.show()
```

```python
var=sns.load_dataset('penguins').head(20)
var
```

```python
sns.scatterplot(x='bill_length_mm', y='bill_depth_mm',data=var, hue='sex')
plt.show()
```

```python
sns.scatterplot(x='bill_length_mm', y='bill_depth_mm',data=var, hue='sex', style='sex')
plt.show()
```

```python
sns.scatterplot(x='bill_length_mm', y='bill_depth_mm',data=var, hue='sex', style='sex', size='sex')
plt.show()
```

```python
sns.scatterplot(x='bill_length_mm', y='bill_depth_mm',data=var, hue='sex', style='sex', size='sex',sizes=(70,170))
plt.show()
```

```python
sns.scatterplot(x='bill_length_mm', y='bill_depth_mm',
                data=var, hue='sex', style='sex', size='sex',sizes=(70,170),palette='s')
plt.show()
```

```python
sns.scatterplot(x='bill_length_mm', y='bill_depth_mm',
                data=var, hue='sex', style='sex', size='sex',sizes=(70,170),palette='cividis_r')
plt.show()
```

```python
sns.scatterplot(x='bill_length_mm', y='bill_depth_mm',
                data=var, hue='sex', style='sex', size='sex',
                sizes=(70,170),palette='cividis_r',alpha=0.5)
plt.show()
```

```python
m={'Male': '*', 'Female': 'o'}
sns.scatterplot(x='bill_length_mm', y='bill_depth_mm',
                data=var, hue='sex', style='sex', size='sex',
                sizes=(70,170),palette='cividis_r',markers=m)
plt.show()
```
