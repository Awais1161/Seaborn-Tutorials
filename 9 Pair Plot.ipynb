```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
var=sns.load_dataset('tips')
var
```

```python
sns.pairplot(var) # only for numeric columns
plt.show()
```

```python
sns.pairplot(var, hue='sex') 
plt.show()
```

```python
sns.pairplot(var,vars=['total_bill', 'tip'], hue='sex') 
plt.show()
```

```python
sns.pairplot(var,vars=['total_bill', 'tip'], hue='sex', hue_order=['Female', 'Male']) 
plt.show()
```

```python
sns.pairplot(var,vars=['total_bill', 'tip'], hue='sex', hue_order=['Female', 'Male'], palette='cubehelix') 
plt.show()
```

```python
sns.pairplot(var, hue='sex', hue_order=['Female', 'Male'], palette='cubehelix', x_vars=['total_bill','tip']) 
plt.show()
```

```python
sns.pairplot(var, hue='sex', hue_order=['Female', 'Male'], palette='cubehelix', y_vars=['total_bill','tip']) 
plt.show()
```

```python
sns.pairplot(var, hue='sex', hue_order=['Female', 'Male'], palette='cubehelix') 
plt.show()
```

```python
sns.pairplot(var, hue='sex', hue_order=['Female', 'Male'], palette='cubehelix', kind='reg') # kind= reg, hist, kde, scatter
plt.show()
```

```python
sns.pairplot(var, hue='sex', hue_order=['Female', 'Male'], palette='cubehelix', kind='kde') # kind= reg, hist, kde, scatter
plt.show()
```

```python
sns.pairplot(var, hue='sex', hue_order=['Female', 'Male'],
             palette='cubehelix', kind='kde', diag_kind='hist') 
plt.show()
```

```python
sns.pairplot(var, hue='sex', hue_order=['Female', 'Male'],markers=['*', '>'])
              
plt.show()
```
